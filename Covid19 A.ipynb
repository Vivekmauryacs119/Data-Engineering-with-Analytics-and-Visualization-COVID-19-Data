```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
file_path = 'WHO-COVID-19-global-hosp-icu-data.csv'
df = pd.read_csv("WHO-COVID-19-global-hosp-icu-data.csv")


df['Date_reported'] = pd.to_datetime(df['Date_reported'], format='%Y-%m-%d', errors='coerce')


col_name = 'Covid_new_hospitalizations_last_7days'
df[col_name] = pd.to_numeric(df[col_name], errors='coerce')


plot_df = df.dropna(subset=['Date_reported', col_name])


grouped = plot_df.groupby(['Date_reported', 'WHO_region'])[col_name].sum().reset_index()


pivot_df = grouped.pivot(index='Date_reported', columns='WHO_region', values=col_name)

plt.figure(figsize=(12, 6))
for region in pivot_df.columns:
    plt.plot(pivot_df.index, pivot_df[region], marker='o', label=region)

plt.xlabel('Date Reported')
plt.ylabel('New COVID-19 Hospitalizations (last 7 days)')
plt.title('Trend of New COVID-19 Hospitalizations in the Last Days by WHO Region')
plt.legend(title='WHO Region')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print('Plot generated successfully.')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("WHO-COVID-19-global-daily-data (1).csv")

df.head()
```

```python
# Check for missing values
df.isnull().sum()

# Fill missing new cases/deaths with 0
df['New_cases'].fillna(0, inplace=True)
df['New_deaths'].fillna(0, inplace=True)

# Convert date column to datetime
df['Date_reported'] = pd.to_datetime(df['Date_reported'])

# Overview of the cleaned data
df.info()
```

```python
# Global cumulative numbers
total_cases = df['New_cases'].sum()
total_deaths = df['New_deaths'].sum()

print(f"Total Global Cases: {int(total_cases):,}")
print(f"Total Global Deaths: {int(total_deaths):,}")
```

```python
latest_data = df[df['Date_reported'] == df['Date_reported'].max()]
top_cases = latest_data.sort_values(by='Cumulative_cases', ascending=False).head(10)
top_deaths = latest_data.sort_values(by='Cumulative_deaths', ascending=False).head(10)

# Plotting
plt.figure(figsize=(14, 6))
sns.barplot(x='Cumulative_cases', y='Country', data=top_cases, palette='viridis')
plt.title("Top 10 Countries by Cumulative Cases")
plt.xlabel("Cases")
plt.ylabel("Country")
plt.show()
```

```python
global_daily = df.groupby('Date_reported')[['New_cases', 'New_deaths']].sum().reset_index()

# Plotting
plt.figure(figsize=(15, 6))
sns.lineplot(data=global_daily, x='Date_reported', y='New_cases', label='New Cases')
sns.lineplot(data=global_daily, x='Date_reported', y='New_deaths', label='New Deaths')
plt.title("Global Daily New Cases and Deaths")
plt.xlabel("Date")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
regional_data = df.groupby(['WHO_region', 'Date_reported'])[['New_cases']].sum().reset_index()

# Plotting cases trend for each region
plt.figure(figsize=(15, 8))
sns.lineplot(data=regional_data, x='Date_reported', y='New_cases', hue='WHO_region')
plt.title("COVID-19 New Cases by WHO Region Over Time")
plt.xlabel("Date")
plt.ylabel("New Cases")
plt.legend(title="WHO Region")
plt.show()
```

```python
# Example: Trend for India
india_data = df[df['Country'] == 'India']

plt.figure(figsize=(15, 6))
sns.lineplot(data=india_data, x='Date_reported', y='New_cases', label='India - New Cases')
sns.lineplot(data=india_data, x='Date_reported', y='New_deaths', label='India - New Deaths')
plt.title("India COVID-19 Daily Cases and Deaths")
plt.xlabel("Date")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
# 1. Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")

# 2. Load the dataset
df = pd.read_csv("WHO-COVID-19-global-data (2).csv")

# 3. Clean and preprocess
df['Date_reported'] = pd.to_datetime(df['Date_reported'])
df['New_cases'].fillna(0, inplace=True)
df['New_deaths'].fillna(0, inplace=True)


print(" Global COVID-19 Overview:")
print(f"Total Cases: {df['New_cases'].sum():,.0f}")
print(f"Total Deaths: {df['New_deaths'].sum():,.0f}")
```

```python
# 5. Plot Global Daily New Cases & Deaths
global_trend = df.groupby('Date_reported')[['New_cases', 'New_deaths']].sum().reset_index()

plt.figure(figsize=(14, 6))
sns.lineplot(data=global_trend, x='Date_reported', y='New_cases', label='New Cases')
sns.lineplot(data=global_trend, x='Date_reported', y='New_deaths', label='New Deaths')
plt.title(' Global Daily New COVID-19 Cases and Deaths')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 6. Top 10 Countries by Cumulative Cases
latest = df[df['Date_reported'] == df['Date_reported'].max()]
top10_cases = latest.sort_values(by='Cumulative_cases', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(y='Country', x='Cumulative_cases', data=top10_cases, palette='magma')
plt.title(" Top 10 Countries by Total Cases")
plt.xlabel("Cumulative Cases")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

```python
# 7. Top 10 Countries by Cumulative Deaths
top10_deaths = latest.sort_values(by='Cumulative_deaths', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(y='Country', x='Cumulative_deaths', data=top10_deaths, palette='rocket')
plt.title(" Top 10 Countries by Total Deaths")
plt.xlabel("Cumulative Deaths")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

```python
# 8. Regional Trend Analysis
regional_trend = df.groupby(['Date_reported', 'WHO_region'])['New_cases'].sum().reset_index()

plt.figure(figsize=(14, 8))
sns.lineplot(data=regional_trend, x='Date_reported', y='New_cases', hue='WHO_region')
plt.title(" New COVID-19 Cases by WHO Region Over Time")
plt.xlabel("Date")
plt.ylabel("New Cases")
plt.legend(title="Region")
plt.tight_layout()
plt.show()
```

```python
# 9. Country-Specific Example (e.g., India)
country_name = "India"
country_df = df[df['Country'] == country_name]

plt.figure(figsize=(14, 6))
sns.lineplot(data=country_df, x='Date_reported', y='New_cases', label='New Cases')
sns.lineplot(data=country_df, x='Date_reported', y='New_deaths', label='New Deaths')
plt.title(f" COVID-19 Trend in {country_name}")
plt.xlabel("Date")
plt.ylabel("Count")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#  Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#  Load Dataset
file_path = "WHO-COVID-19-global-hosp-icu-data.csv"
df = pd.read_csv(file_path)

#  Data Cleaning
df['Date_reported'] = pd.to_datetime(df['Date_reported'])

# Drop rows where all hospitalization and ICU columns are NaN
df.dropna(subset=[
    'Covid_new_hospitalizations_last_7days',
    'Covid_new_icu_admissions_last_7days',
    'Covid_new_hospitalizations_last_28days',
    'Covid_new_icu_admissions_last_28days'
], how='all', inplace=True)

#  Basic Info
print("Dataset Info:")
print(df.info())
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
#  Time Trend of Global Hospitalizations (28 days)
df_grouped_date = df.groupby('Date_reported')[
    'Covid_new_hospitalizations_last_28days'
].sum().reset_index()

plt.figure(figsize=(14,6))
sns.lineplot(data=df_grouped_date, x='Date_reported', y='Covid_new_hospitalizations_last_28days')
plt.title("Global COVID-19 Hospitalizations Over Time (Last 28 Days)")
plt.xlabel("Date")
plt.ylabel("Total Hospitalizations")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#  Top 10 Countries by Hospitalizations (28 Days)
top_countries_hosp = df.groupby("Country")[
    'Covid_new_hospitalizations_last_28days'
].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_countries_hosp.values, y=top_countries_hosp.index, palette="coolwarm")
plt.title("Top 10 Countries by COVID-19 Hospitalizations (Last 28 Days)")
plt.xlabel("Total Hospitalizations")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

```python
#  ICU Admissions Trend for a Sample Country (e.g., Italy)
italy_data = df[df["Country"] == "ITALY"]
plt.figure(figsize=(14,6))
sns.lineplot(data=italy_data, x='Date_reported', y='Covid_new_icu_admissions_last_7days')
plt.title("Italy: COVID-19 ICU Admissions Over Time (7 Days)")
plt.xlabel("Date")
plt.ylabel("ICU Admissions")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#  Hospitalization by WHO Region
region_data = df.groupby("WHO_region")[
    'Covid_new_hospitalizations_last_28days'
].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=region_data.values, y=region_data.index, palette="viridis")
plt.title("Total COVID-19 Hospitalizations by WHO Region (Last 28 Days)")
plt.xlabel("Total Hospitalizations")
plt.ylabel("WHO Region")
plt.tight_layout()
plt.show()

#  Insights
print("\n Insights:")
print("- Hospitalization trends vary greatly by country and region.")
print("- Some countries report consistently high numbers over time.")
print("- ICU admissions appear less frequently reported than hospitalizations.")
print("- WHO European and Americas regions have higher overall hospitalization counts.")

#  Conclusion
print("\n Conclusion:")
print("This analysis provided a high-level view of COVID-19 hospitalization and ICU trends worldwide.")
print("The data suggests variation in reporting and healthcare burden across regions and time.")
```

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("WHO-COVID-19-global-monthly-death-by-age-data.csv")

# Preview
df.head()
```

```python
# Clean column names
df.columns = df.columns.str.strip().str.upper().str.replace(" ", "_")

# Drop rows with missing death counts
df = df.dropna(subset=["DEATHS"])

# Convert DEATHS to numeric
df["DEATHS"] = pd.to_numeric(df["DEATHS"], errors="coerce")

# Check for nulls
df.isnull().sum()
```

```python

df.groupby("AGEGROUP")["DEATHS"].sum().sort_values(ascending=False).plot(kind="bar", title="Total Deaths by Age Group")
plt.ylabel("Deaths")
plt.show()



df.groupby("WHO_REGION")["DEATHS"].sum().sort_values(ascending=False).plot(kind="bar", color="orange", title="Total Deaths by WHO Region")
plt.ylabel("Deaths")
plt.show()


df.groupby("WB_INCOME")["DEATHS"].sum().sort_values(ascending=False).plot(kind="bar", color="green", title="Total Deaths by Income Group")
plt.ylabel("Deaths")
plt.show()
```

```python

monthly = df.groupby(["YEAR", "MONTH"])["DEATHS"].sum().reset_index()
monthly["DATE"] = pd.to_datetime(monthly["YEAR"].astype(str) + "-" + monthly["MONTH"].astype(str) + "-01")

plt.figure(figsize=(12, 5))
sns.lineplot(data=monthly, x="DATE", y="DEATHS")
plt.title("Monthly Deaths Over Time")
plt.xlabel("Date")
plt.ylabel("Deaths")
plt.show()
```
